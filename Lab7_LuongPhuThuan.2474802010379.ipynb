```python

# 1. IMPORT THƯ VIỆN

import numpy as np
# 2. TÍCH VECTOR ĐƠN GIẢN
print("== Phần 1: Tích vector ==")
signals = np.array([[1, 2], [3, 4], [10, 0]])
sample = np.array([3, -1])

result_inner = np.inner(signals, sample)
print("np.inner(signals, sample):", result_inner)  # [1 5 30]

# 3. TÌM ÂM THANH CON TRONG CHUỖI ÂM
print("\n== Phần 2: Tìm chuỗi âm thanh con ==")
A = np.array([0, 0, -1, 2, 3, -1, 0, 1, -1, -1])
search_vector = np.array([0, 1, -1])

print("Độ dài chuỗi A:", len(A))
print("Độ dài search_vector:", len(search_vector))

# Tạo ma trận B gồm các đoạn con (window) từ A
rows = len(A) - len(search_vector) + 1
cols = len(search_vector)
B = np.zeros((rows, cols))

for i in range(rows):
    for j in range(cols):
        B[i, j] = A[i + j]

print("\nMa trận trượt B:")
print(B.astype(int))  # hiển thị dưới dạng số nguyên

# Tính dot product giữa từng hàng của B và search_vector
C = np.inner(B, search_vector)
print("\nVector kết quả dot product C:")
print(C.astype(int))

# Giá trị cần tìm
target = np.inner(search_vector, search_vector)
print("\nGiá trị cần so khớp:", target)

# Tìm vị trí khớp
print("\nVị trí tìm thấy:")
for i in range(rows):
    if C[i] == target:
        print(f"Vị trí {i}: {B[i].astype(int)}")
```

```python
from PIL import Image
import requests
from io import BytesIO


# 1. Tải ảnh mẫu từ Internet (vì bạn không thể mở file từ ổ C: trên Colab)

url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/e/eb/The_Earth_seen_from_Apollo_17.jpg/600px-The_Earth_seen_from_Apollo_17.jpg'
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# 2. Thông tin ảnh

print("Chiều rộng:", img.width)
print("Chiều cao:", img.height)
print("Kiểu ảnh:", img.mode)

# Nếu ảnh không phải RGB, chuyển về RGB
if img.mode != "RGB":
    img = img.convert("RGB")
    print("Đã chuyển ảnh về dạng RGB.")


# 3. Thu nhỏ ảnh còn 1/2 kích thước

new_width = int(img.width / 2)
new_height = int(img.height / 2)
new_img = img.resize((new_width, new_height), Image.ANTIALIAS)


# 4. Hiển thị ảnh gốc và ảnh thu nhỏ

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Ảnh gốc")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(new_img)
plt.title("Ảnh đã thu nhỏ")
plt.axis('off')

plt.show()
# 5. Lưu ảnh thu nhỏ thành file mới
new_img.save("earth_small.jpg")
print("Đã lưu ảnh mới thành earth_small.jpg")
```

```python
# 1. Cài thư viện
from PIL import Image, ImageDraw
import numpy as np

# 2. Mở ảnh đã upload (đổi tên file nếu cần)
input_image = Image.open('/mnt/data/8650eaed-6029-40c6-8538-67930ac8e0a4.png').convert("RGB")
input_pixels = input_image.load()

# 3. Tạo kernel làm mờ trung bình 3x3
box_kernel = [
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
]
kernel = box_kernel
offset = len(kernel) // 2

# 4. Tạo ảnh mới để lưu kết quả
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# 5. Áp dụng bộ lọc làm mờ (convolution)
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]  # [R, G, B]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))

# 6. Lưu ảnh kết quả
output_path = "/mnt/data/traidat_lammo.png"
output_image.save(output_path)
print("Đã lưu ảnh sau khi làm mờ tại:", output_path)

# 7. Hiển thị ảnh trước và sau
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title("Ảnh gốc")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(output_image)
plt.title("Ảnh đã làm mờ")
plt.axis("off")

plt.show()
```

```python
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt

# 1. Mở ảnh gốc bạn đã upload
input_image = Image.open('/mnt/data/8650eaed-6029-40c6-8538-67930ac8e0a4.png').convert("RGB")
input_pixels = input_image.load()

# 2. Gaussian kernel 5x5
gauss_kernel = [
    [1,  4,  6,  4, 1],
    [4, 16, 24, 16, 4],
    [6, 24, 36, 24, 6],
    [4, 16, 24, 16, 4],
    [1,  4,  6,  4, 1]
]
# Chuẩn hóa kernel bằng cách chia cho 256
kernel = [[value / 256 for value in row] for row in gauss_kernel]
offset = len(kernel) // 2

# 3. Tạo ảnh mới để lưu kết quả
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

# 4. Áp dụng Gaussian blur
for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))

# 5. Lưu và hiển thị ảnh kết quả
output_path = "/mnt/data/traidat_gauss.png"
output_image.save(output_path)
print("Đã lưu ảnh làm mờ Gaussian tại:", output_path)

# 6. So sánh ảnh trước và sau
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title("Ảnh gốc")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(output_image)
plt.title("Ảnh làm mờ Gauss")
plt.axis("off")

plt.show()
```

```python
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt

# Mở ảnh gốc đã upload
input_image = Image.open('/mnt/data/8650eaed-6029-40c6-8538-67930ac8e0a4.png').convert("RGB")
input_pixels = input_image.load()

# 2 kernel rút xương
kernels = {
    "Kernel A": [
        [0, -1, 0],
        [-1, 4, -1],
        [0, -1, 0]],
    "Kernel B": [
        [-1, -1, -1],
        [-1, 8, -1],
        [-1, -1, -1]]}

def apply_kernel(input_image, kernel):
    input_pixels = input_image.load()
    offset = len(kernel) // 2
    output_image = Image.new("RGB", input_image.size)
    draw = ImageDraw.Draw(output_image)

    for x in range(offset, input_image.width - offset):
        for y in range(offset, input_image.height - offset):
            acc = [0, 0, 0]
            for a in range(len(kernel)):
                for b in range(len(kernel)):
                    xn = x + a - offset
                    yn = y + b - offset
                    pixel = input_pixels[xn, yn]
                    acc[0] += pixel[0] * kernel[a][b]
                    acc[1] += pixel[1] * kernel[a][b]
                    acc[2] += pixel[2] * kernel[a][b]

            # Cắt giới hạn trong khoảng [0, 255]
            r = min(max(int(acc[0]), 0), 255)
            g = min(max(int(acc[1]), 0), 255)
            b = min(max(int(acc[2]), 0), 255)
            draw.point((x, y), (r, g, b))

    return output_image

# Áp dụng từng kernel và hiển thị kết quả
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(input_image)
plt.title("Ảnh gốc")
plt.axis("off")

for i, (label, kernel) in enumerate(kernels.items()):
    result = apply_kernel(input_image, kernel)
    result.save(f"/mnt/data/traidat_{label.replace(' ', '_').lower()}.png")
    plt.subplot(1, 3, i + 2)
    plt.imshow(result)
    plt.title(f"Kết quả với {label}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
