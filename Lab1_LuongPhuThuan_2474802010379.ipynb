```python
# Nhập thư viện cần thiết
from sympy import Symbol, solve, sqrt

# Định nghĩa biến
x = Symbol('x')

# ----------------------------
# Ví dụ 1: Giải phương trình x + 3 - 5 = 0
bieuthuc1 = x + 3 - 5
nghiem1 = solve(bieuthuc1)
print("Nghiệm của phương trình x + 3 - 5 = 0 là:", nghiem1)
# Kết quả: [2]

# ----------------------------
# Ví dụ 2: Giải phương trình x^2 + 3 - 5 = 0
bieuthuc2 = x**2 + 3 - 5
nghiem2 = solve(bieuthuc2)
print("Nghiệm của phương trình x^2 + 3 - 5 = 0 là:", nghiem2)
# Kết quả: [-sqrt(2), sqrt(2)]

# Truy cập từng nghiệm
print("Nghiệm thứ nhất:", nghiem2[0])  # -sqrt(2)
print("Nghiệm thứ hai:", nghiem2[1])   # sqrt(2)
```

```python
from sympy import Symbol, solve, pprint

# Khai báo các biến x, a, b, c
x = Symbol('x')
a = Symbol('a')
b = Symbol('b')
c = Symbol('c')

# Định nghĩa phương trình tổng quát bậc 2
ptb2 = a*x**2 + b*x + c

# Giải phương trình theo biến x, với nghiệm trả về là dict
nghiem = solve(ptb2, x, dict=True)

# In kết quả
print("Nghiệm tổng quát của phương trình ax^2 + bx + c = 0 là:")
pprint(nghiem)
```

```python
from sympy import Symbol, solve

# Khai báo các biến x và y
x = Symbol('x')
y = Symbol('y')

# Định nghĩa 2 phương trình
pt1 = 2*x + 3*y - 12
pt2 = 3*x - 2*y - 5

# Giải hệ phương trình
nghiem = solve((pt1, pt2), dict=True)

# In nghiệm đầu tiên (vì có thể có nhiều nghiệm)
print("Nghiệm của hệ phương trình là:")
print(nghiem)

# Lấy nghiệm đầu tiên (vì solve trả về danh sách dict)
nghiem_xy = nghiem[0]

# Thử lại nghiệm bằng cách thay vào phương trình ban đầu
print("\nThử lại nghiệm bằng cách thay vào pt1:")
print(pt1.subs({x: nghiem_xy[x], y: nghiem_xy[y]}))

print("\nThử lại nghiệm bằng cách thay vào pt2:")
print(pt2.subs({x: nghiem_xy[x], y: nghiem_xy[y]}))
```

```python
import numpy as np

# Khai báo ma trận M1
M1 = np.array([[9, 12], [23, 30]])
print("Ma trận M1:")
print(M1)

# Tạo vector hàng u = (2,1)
u = np.array([2, 1])

# Tích M1 · u
tichM1u = M1.dot(u)
print("\nTích M1 · u:")
print(tichM1u)

# Hoặc dùng np.dot
tichM1u_alt = np.dot(M1, u)
print("\nTích M1 · u dùng np.dot(M1, u):")
print(tichM1u_alt)

# Dùng np.dot(u, M1)
tich_uM1 = np.dot(u, M1)
print("\nTích u · M1:")
print(tich_uM1)

# ========================
# Các thao tác thực hành tiếp theo

# Tạo ma trận mat1 là ma trận toàn 0 với kích thước 5x5
mat1 = np.zeros((5,5))
print("\nMa trận mat1:")
print(mat1)

# Tạo ma trận mat2 toàn số 1 kích thước 5x5
mat2 = np.ones((5,5))
print("\nMa trận mat2:")
print(mat2)

# Tính mat3 = mat1 + 2*mat2
mat3 = mat1 + 2 * mat2
print("\nMa trận mat3 = mat1 + 2*mat2:")
print(mat3)

# Gán mat4 = mat3 (tham chiếu, không sao chép)
mat4 = mat3

# Thay đổi một phần tử của mat4
mat3[1][1] = 10
print("\nSau khi thay đổi mat3[1][1] = 10:")
print("mat3:")
print(mat3)
print("mat4 (do tham chiếu nên cũng thay đổi):")
print(mat4)

# Sao chép mat3 thành mat5 độc lập
mat5 = np.copy(mat3)

# Thay đổi phần tử mat3[1][2] = 10
mat3[1][2] = 10
print("\nSau khi thay đổi mat3[1][2] = 10:")
print("mat3:")
print(mat3)
print("mat5 (do copy nên KHÔNG thay đổi):")
print(mat5)
```

```python
import numpy as np

# Các ma trận nguy cơ
A = np.array([[1, 0, 0, 0, 1],
              [3, 1, 0, 1, 2],
              [5, 2, 0, 1, 2],
              [2, 0, 1, 2, 3]])

B = np.array([[1, 2, 2, 2, 0],
              [0, 1, 2, 2, 4],
              [0, 1, 2, 4, 2],
              [1, 4, 1, 2, 0]])

C = np.array([[0, 1, 1, 1, 3],
              [0, 0, 1, 1, 4],
              [0, 0, 1, 3, 3],
              [1, 1, 3, 3, 0]])

D = np.array([[3, 1, 1, 0, 1],
              [7, 0, 0, 3, 5],
              [5, 0, 0, 3, 5],
              [0, 0, 3, 0, 5]])

E = np.array([[0, 0, 0, 0, 0],
              [0, 5, 15, 5, 0],
              [0, 5, 15, 5, 0],
              [20, 5, 0, 5, 0]])

# Tính tổng nguy cơ theo từng yêu cầu
# a. Tránh lộ bí mật => chỉ xét E <= 5
print("\n(a) An toàn trong chiến dịch ngắn hạn 1-2 ngày (tránh lộ bí mật):")
result_a = (E <= 5)
print(result_a.astype(int))

# b. Tránh lộ bí mật và bệnh dịch => E <= 5 và D <= 3
print("\n(b) An toàn trong tập luyện thời bình (tránh lộ bí mật và bệnh dịch):")
result_b = (E <= 5) & (D <= 3)
print(result_b.astype(int))

# c. Tránh cháy rừng và bệnh dịch => A <= 2 và D <= 3
print("\n(c) An toàn theo mùa khô (tránh cháy rừng và bệnh dịch):")
result_c = (A <= 2) & (D <= 3)
print(result_c.astype(int))

# d. Tránh lũ, lở núi và bệnh dịch => B <= 2, C <= 2, D <= 3
print("\n(d) An toàn trong mùa mưa (tránh lũ, sạt lở, bệnh dịch):")
result_d = (B <= 2) & (C <= 2) & (D <= 3)
print(result_d.astype(int))

# e. Chỉ tránh lộ bí mật => E <= 5
print("\n(e) An toàn trong 8 tháng (chỉ cần tránh lộ bí mật):")
result_e = (E <= 5)
print(result_e.astype(int))
```
